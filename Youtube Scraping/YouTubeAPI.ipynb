```python
def find_comments(vidId):
    try :
        req2 = youtube.commentThreads().list(part="snippet, replies", videoId=vidId)
        res2 = req2.execute()
        for item in res2['items']:
            print (item['snippet']['topLevelComment']['snippet']['textOriginal'] )
            print("-------")
    except :
        print("not connected")
    
    
```
